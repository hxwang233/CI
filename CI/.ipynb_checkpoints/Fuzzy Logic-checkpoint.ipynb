```python
import pandas as pd 
import numpy as np
import itertools
import matplotlib.pyplot as plt
```

```python
rules  = pd.read_table("../dataSet/覆冰规则.txt", sep='\\s+', engine='python')
data   = pd.read_excel("../dataSet/覆冰验证数据.xlsx")
data   = np.array(data)
x_data = data[:,0:4]
y_data = data[:,4]
rule = dict()
for i in rules.values:
    k = tuple(i[0:4])
    if k not in rule.keys():
        rule[k] = set()
    rule[k].add(i[4])
```

```python
"""隶属度函数"""
class ET_Subordinate_func():
    def NB(self, x):
        if x >= -20 and x <= -10:
            return -0.1 * x - 1
        else:
            return 0
    def NS(self, x):
        if x >= -20 and x < -10:
            return 0.1 * x + 2
        elif x >= -10 and x <= 0:
            return -0.1 * x
        else:
            return 0
    def O(self, x):
        if x >= -10 and x < 0:
            return 0.1 * x + 1
        elif x >=0 and x <= 10:
            return -0.1 * x + 1
        else:
            return 0
    def PS(self, x):
        if x >= 0 and x < 10:
            return 0.1 * x
        elif x >= 10 and x <= 20:
            return -0.1 * x + 2
        else:
            return 0
    def PB(self, x):
        if x >= 10 and x <= 20:
            return 0.1 * x - 1
        else:
            return 0
    def subordinate(self, x):
        res = dict()
        res['NB'] = self.NB(x)
        res['NS'] = self.NS(x)
        res['O']  = self.O(x)
        res['PS'] = self.PS(x)
        res['PB'] = self.PB(x)
        return res # 环境温度  （同 导线温度）

class EH_Subordinate_func(ET_Subordinate_func):
    def NB(self,x):
        if x >= 0 and x <= 25:
            return -1/25 * x + 1
        else:
            return 0
    def NS(self,x):
        if x >= 0 and x < 25:
            return 1/25 * x
        elif x >= 25 and x <= 50:
            return -1/25 * x + 2 
        else:
            return 0
    def O(self,x):
        if x >= 25 and x < 50:
            return 1/25 * x - 1
        elif x >= 50 and x <= 75:
            return -1/25 * x + 3
        else:
            return 0
    def PS(self,x):
        if x >= 50 and x < 75:
            return 1/25 * x - 2
        elif x >= 75 and x <= 100:
            return -1/25 * x + 4
        else:
            return 0
    def PB(self,x):
        if x >= 75 and x <= 100:
            return 1/25 * x - 3
        else:
            return 0  # 环境湿度
    
class EW_Subordinate_func(ET_Subordinate_func):
    def NB(self,x):
        if x >= 0 and x <= 5:
            return -1/5 * x + 1
        else:
            return 0
    def NS(self,x):
        if x >= 0 and x < 5:
            return 1/5 * x
        elif x >= 5 and x <= 10:
            return -1/5 * x + 2 
        else:
            return 0
    def O(self,x):
        if x >= 5 and x < 10:
            return 1/5 * x - 1
        elif x >= 10 and x <= 15:
            return -1/5 * x + 3
        else:
            return 0
    def PS(self,x):
        if x >= 10 and x < 15:
            return 1/5 * x - 2
        elif x >= 15 and x <= 20:
            return -1/5 * x + 4
        else:
            return 0
    def PB(self,x):
        if x >= 15 and x <= 20:
            return 1/5 * x - 3
        else:
            return 0  # 环境风速

class IT_Subordinate_func(ET_Subordinate_func):
    def NB(self,x):
        if x >= 0 and x <= 7.5:
            return -1/7.5 * x + 1
        else:
            return 0
    def NS(self,x):
        if x >= 0 and x < 7.5:
            return 1/7.5 * x
        elif x >= 7.5 and x <= 15:
            return -1/7.5 * x + 2 
        else:
            return 0
    def O(self,x):
        if x >= 7.5 and x < 15:
            return 1/7.5 * x - 1
        elif x >= 15 and x <= 22.5:
            return -1/7.5 * x + 3
        else:
            return 0
    def PS(self,x):
        if x >= 15 and x < 22.5:
            return 1/7.5 * x - 2
        elif x >= 22.5 and x <= 30:
            return -1/7.5 * x + 4
        else:
            return 0
    def PB(self,x):
        if x >= 22.5 and x <= 30:
            return 1/7.5 * x - 3
        else:
            return 0  # 覆冰厚度
```

```python
et = ET_Subordinate_func()
eh = EH_Subordinate_func()
ew = EW_Subordinate_func()
ct = ET_Subordinate_func()
it = IT_Subordinate_func()
calculators = [et, eh, ew, ct, it]
estimate = list()
for record in x_data:  # 遍历训练样本
    res = 0
    weight = list()
    condition = dict()
    subo = list()
    for i,j in enumerate(record):
        temp = calculators[i].subordinate(j)  # 模糊化
        delIndex = list()
        '''去除隶属度为0的'''
        for k,v in temp.items():
            if v == 0: 
                delIndex.append(k)
        for k in delIndex:
            del temp[k]
        subo.append(temp)
    #print(subo)
    #笛卡尔乘积
    for k1,v1 in subo[0].items():
        for k2,v2 in subo[1].items():
            for k3,v3 in subo[2].items():
                for k4,v4 in subo[3].items():
                    if (k1,k2,k3,k4) in rule.keys(): # 激活规则
                        condition[(k1,k2,k3,k4)] = min((v1,v2,v3,v4)) #取最小
    conclusion = dict()
    #print(condition)
    for i,j in condition.items():
        for c in rule[i]:
            if c not in conclusion.keys():
                conclusion[c] = list()
            conclusion[c].append(j)
            break
    if len(conclusion) > 2:   
        if max(conclusion['O']) >= max(conclusion['NB']):
               del conclusion['O']
        else: 
               del conclusion['NB']
    print(conclusion)
    for c,v in conclusion.items():
        w = max(v)  #取最大
        weight.append(w)
        if c == 'NB':
            res += 3.75 * w
        elif c == 'NS':
            res += 7.5 * w
        elif c == 'O':
            res += 15 * w
        elif c == 'PS': 
            res += 22.5 * w
        elif c == 'PB':
            res += 26.25 * w
    estimate.append(res / sum(weight))  #去模糊化
print(estimate)
```

```python
x = list(range(1,17))
plt.ylim(0,20)
plt.scatter(x,estimate)
plt.plot(x,estimate)
plt.scatter(x,y_data)
plt.plot(x,y_data)
plt.legend(['result','real'])
plt.savefig("各样本点实际覆冰与预测覆冰数据的对比 .png")
plt.show()
```
